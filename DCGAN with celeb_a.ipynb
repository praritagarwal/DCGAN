## Loading preprocessed celeba images into a tensorflow dataset

This is based on tensorflow's [_Load images_ tutorial](https://www.tensorflow.org/tutorials/load_data/images#load_using_tfdata) 

```python
# glob pattern for the image paths
celeba_filepaths = 'processed_celeba_small/celeba/*.jpg'

# first we have to create a dataset of filepaths from the above glob pattern
filepaths_dataset = tf.data.Dataset.list_files(celeba_filepaths)
```

```python
# test a random set of images to check that the filepaths were generated correctly
from PIL import Image

imgs = []
for path in filepaths_dataset.take(20):
    img = Image.open(path.numpy().decode()) # use decode to convert from binary string to ordinary string
    imgs.append(img)

# display the sample images
fig, ax = plt.subplots(figsize = (20,4), nrows = 2, ncols = 10)
ctr = 0
for row in range(2):
    for col in range(10):
        ax[row, col].imshow(imgs[ctr])
        ax[row, col].set_xticks([])
        ax[row, col].set_yticks([])
        ctr+=1
plt.show()        
```

```python
# function to read, decode and rescale images from their filepath

def read_decode_and_rescale(imagepath, rescale = True, min_ = -1, max_ = 1):
    image = tf.io.read_file(imagepath)
    image = tf.image.decode_jpeg(image, channels = 3 ) 
    
    if not rescale:
        return image

    image = (max_-min_)*(image/255) + min_ 
    return image
```

```python
# function to restore the scale of the image to lie between 0 and 1
def restore_scale(image, min_ = -1, max_ = 1):
    ''' min_: minimum pixel value expected in the input image
        max_: maximum pixel value expected in the input image'''
    
    image = (image-min_)/(max_-min_)
    return image
```

```python
# apply read_decode_and_rescale to filepaths_dataset
num_parallel_calls = tf.data.experimental.AUTOTUNE
images_dataset = filepaths_dataset.map(read_decode_and_rescale, 
                                       num_parallel_calls = num_parallel_calls)
```

```python
# sample a batch of images from images_dataset
sample = iter(images_dataset.take(20))

# display the images in the sample
fig, ax = plt.subplots(figsize = (20, 4), nrows = 2, ncols = 10)
for row in range(2):
    for col in range(10):
        img = restore_scale(next(sample))
        ax[row, col].imshow(img.numpy())
        ax[row, col].set_xticks([])
        ax[row, col].set_yticks([])
plt.show()        
```

## Build a DCGAN

```python
from my_DCGAN_class import DCGAN
```

```python
dcgan = DCGAN(restore_scale)
```

### Generator

```python
# generator architecture
coding_size = 100
l1 = (4*4*1024, 'selu')
l2 = (4, 4)
l3 = (512, 3, 2, 'same', 'selu')
l4 = (256, 3, 2, 'same', 'selu')
l5 = (128, 3, 2, 'same', 'selu')
l6 = (64, 3, 2, 'same', 'selu')
l7 = (3, 1, 1, 'same', 'tanh')
kernel_initializer = tf.keras.initializers.RandomNormal(stddev = 0.02)
list_layers = [l1,l2,l3,l4,l5,l6, l7]

dcgan.build_generator(list_layers, kernel_initializer = kernel_initializer)
dcgan.generator.summary()
```

### Discriminator

```python
# discriminator architecture
l1 = (64, 5, 2, 'same', keras.layers.LeakyReLU(0.2))
l2 = (128, 5, 2, 'same', keras.layers.LeakyReLU(0.2))
l3 = (256, 5, 2, 'same', keras.layers.LeakyReLU(0.2))
l4 = (512, 5, 2, 'same', keras.layers.LeakyReLU(0.2))
l5 = (512, 5, 2, 'same', keras.layers.LeakyReLU(0.2))
l6 = (1024, 5, 2, 'same', keras.layers.LeakyReLU(0.2))
kernel_initializer = tf.keras.initializers.RandomNormal(stddev = 0.02)
list_layers = [l1,l2,l3,l4,l5,l6]
input_shape = [64,64,3]
dcgan.build_discriminator(list_layers, input_shape, 
                            kernel_initializer = kernel_initializer)
dcgan.discriminator.summary()
```

### GAN 

```python
optimizer = keras.optimizers.Adam(learning_rate = 0.0002, beta_1 = 0.5)
dcgan.build_GAN(optimizer = optimizer)
dcgan.GAN.summary()
```

### Training the GAN

```python
# creating a batched dataset of true images
batch_size = 32
prefetch = tf.data.experimental.AUTOTUNE
images_dataset = images_dataset.batch(batch_size, drop_remainder = True).prefetch(prefetch)

# training the gan
num_epochs = 50
dcgan.train_gan(training_images = images_dataset, 
                num_epochs = num_epochs, 
                batch_size = batch_size,
                coding_size = coding_size, 
                num_plots = 10)
```
